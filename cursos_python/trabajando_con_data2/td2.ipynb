## Pivotar Tablas

```python
## importar testing de pandas
import pandas.util.testing as tm; tm.N = 3

## Crear una unpivot funci√≥n 
def unpivot(frame):
    N,K = frame.shape
    
    data = {
        'value': frame.values.ravel('F'),
        'variables': np.asanyarray(frame.columns).repeat(N),
        'date':np.tile(np.asarray(frame.index), K)
    }
    
    ## Devuelve el df
    return DataFrame(data, columns=['date', 'variables', 'value'])

# Set del df que usaremos
df = unpivot(tm.makeTimeDataFrame())
```

```python
df
```

```python
df_piv = df.pivot('date', 'variables', 'value')
```

```python
df_piv
```
