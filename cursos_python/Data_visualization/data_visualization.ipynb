## Lección 49 Kernel Density Estimation Plots

```python
dataset = randn(25)
```

```python
sns.rugplot(dataset)
plt.ylim(0,1)
```

```python
plt.hist(dataset, alpha= .5, edgecolor="black")
sns.rugplot(dataset)
```

```python
sns.rugplot(dataset)
x_min = dataset.min() - 2
x_max = dataset.max() + 2

x_axis = np.linspace(x_min, x_max, 100)

bandwidth = (4*dataset.std()**5 / (3*len(dataset))) ** 0.2
kernel_list = []

for data_point in dataset:
    
    ## Crear un bucle kernel para cada pundo y apendizarlo al kernel list
    kernel = stats.norm(data_point, bandwidth).pdf(x_axis)
    kernel_list.append(kernel)
    
    ## Escalamos los gráficos
    kernel = kernel / kernel.max()
    kernel = kernel * 0.4
    
    plt.plot(x_axis, kernel, color='gray', alpha=0.5)

plt.ylim(0,1)
```

```python
sum_of_kde = np.sum(kernel_list, axis=0)

fig = plt.plot(x_axis, sum_of_kde, color='red')

sns. rugplot(dataset)

plt.yticks([])

plt.subtitle("Sum of the basis finctions")
```

```python
sns.kdeplot(dataset)
```

```python
sns.rugplot(dataset, color='black')
for bw in np.arange(0.5, 2, 0.25):
    sns.kdeplot(dataset, bw=bw, lw=1.8, label=bw, shade=True)
```

```python
kernel_options = ['biw', 'cos', 'epa', 'gau', 'tri', 'triw']

for kern in kernel_options:
    sns.kdeplot(dataset, kernel = kern, label=kern)
```

## Lección 53 heatmaps y cluster matrices

```python
df_vuelos = sns.load_dataset('flights')
df_vuelos
```

```python
df_vuelos.to_csv("df_vuelos.csv")
```

```python
df_vuelos = df_vuelos.pivot('month', 'year', 'passengers')
```

```python
df_vuelos
```

```python
sns.heatmap(df_vuelos, annot=True, fmt='d',
            center=df_vuelos.loc['Jan',1955])
```

```python
f, (axis1, axis2) = plt.subplots(2,1)

vuelos_anuales = df_vuelos.sum()
vuelos_anuales

anios = pd.Series(vuelos_anuales.index.values)
anios = pd.DataFrame(anios)

vuelos = pd.Series(vuelos_anuales.values)
vuelos = pd.DataFrame(vuelos) 

df_anios=pd.concat((anios, vuelos), axis=1)
df_anios.columns = ["Años", "Vuelos"]

sns.barplot(x='Años', y="Vuelos",
            data=df_anios, ax=axis1)

sns.heatmap(df_vuelos, cmap="Blues", ax=axis2, cbar_kws={'orientation':'horizontal'})
```

```python
sns.clustermap(df_vuelos)
```

```python
sns.clustermap(df_vuelos,standard_scale=1) ## Cambiamos por 0 para las filas
```
